### **FIFTH STEP: FEATURE ENGINEERING + OHC (TRAIN):**

**OHC FOR CHOSEN TRAIN CATEGORICAL VARIABLES**

Note we chose to only isolate a few categorical variables (Bsmt Cond, Kitchen Qual, and Pool QC) for OHC due to our hypothesis that they these categorical variables focused on 'quality'/'conditions' will have a stronger influence on our target variable, sales price. And we would like them to be OHC to be used in our feature engineering.

We determined that the remaining categorical variables are not worthwhile to do OHC and thus making our data set significantly larger because we predict they will not have a strong influence on the target.

```python
#Building the OHC for our key categorical variables that are are interested in; BsmtCond, KitchenQual, and Pool QC.

from sklearn.preprocessing import OneHotEncoder

# We define our One Hot Encoder
ohe = OneHotEncoder(categories='auto')

# We can use here the qualitative variables that we hypothesize are explanatory variables and most important for OHC
categorical_features = ['Bsmt Cond', 'Kitchen Qual','Pool QC']

# We apply the encoding to our train dataset
feature_arr = ohe.fit_transform(train_reduced[categorical_features]).toarray()

# We also store the names of the categories for the new column names
feature_labels = ohe.categories_

# We apply the new feature names
features = pd.DataFrame(feature_arr, columns=ohe.get_feature_names_out())

# And then combine with the rest of the numerical variables
train_reduced = pd.concat([train_reduced, features], axis=1).drop(columns=categorical_features,
                                                        axis=1)
# Print the head to determine if the correct OHC was added
train_reduced.head()
```

**FEATURE ENG #1: TOTAL PORCH SQUARE FEET**

We analyzed the data to see where there were individual SF metrics but not a total SF and we found that for the 'Porch related' variables. Since we are planning to add up Porch SF in to find a total SF for the property we thought this would be an important one to feature eng for efficiency. Note: this is working under the assumption that some homes have multiple porch types.

```python
#FeatureEng#1: Total Porch SF (combining all SF related features for porches to create one SF for porches)
TotalPorchSF = train_reduced['Wood Deck SF'] + train_reduced['Open Porch SF'] + train_reduced['Enclosed Porch'] + train_reduced['Screen Porch']

# Check to see if feature eng is working
print(TotalPorchSF)

# Rename the Series to 'TotalPorchSF'
TotalPorchSF = TotalPorchSF.rename('TotalPorchSF')

# Concatenate the new TotalPorchSF column with the original DataFrame
train_reduced = pd.concat([train_reduced, TotalPorchSF], axis=1)


# Check to see if the adjusted train data set has the new TotalSqFt Column
print(train_reduced)



```

**FEATURE ENG #2: TOTAL SQUARE FEET**

Next we wanted to build a new feature that would help us determine the entire size of the property. Our team hypothesizes that the entire property size could have a strong influence on the target variable. We leveraged all of our SF (square feet) based variables to understand the entire property size to build our 'TotalSqFt' variable.

```python
#FeatureEng #2: TotalSqFt (Evaluating the total size and livable space)
TotalSqFt = train_reduced['Gr Liv Area'] + train_reduced['Total Bsmt SF'] + train_reduced['TotalPorchSF'] + train_reduced['Garage Area']

# Check to see if feature eng is working
print(TotalSqFt)

# Rename the Series to 'TotalSqFt'
TotalSqFt = TotalSqFt.rename('TotalSqFt')

# Concatenate the new TotalSqFt column with the original DataFrame
train_reduced = pd.concat([train_reduced, TotalSqFt], axis=1)


# Check to see if the adjusted train data set has the new TotalSqFt Column
print(train_reduced)
```

**FEATURE ENG #3: LUXURY RATING**
Next we built this new 'luxury rating' variable. We felt that understanding the 'luxury' of the property - the condition, quality of key features of the home, could be a high predictor of the sales price. Particularly we know based on the earlier correlation analysis that the feature - 'Overall Qual' has a high influence on the 'Sales Price'

```python
#FeatureEng #3: LuxRating (Bringing together key factors that could indicate a more high-end property including quality and additional amenities such as access to a finished pool and basement)
LuxRating = train_reduced['Overall Qual'].fillna(0) + train_reduced['Kitchen Qual_Ex'].fillna(0) + train_reduced['Kitchen Qual_Fa'].fillna(0) + train_reduced['Kitchen Qual_Gd'].fillna(0) + train_reduced['Kitchen Qual_Po'].fillna(0) + train_reduced['Kitchen Qual_TA'].fillna(0) + train_reduced['Pool QC_Ex'].fillna(0) + train_reduced['Pool QC_Fa'].fillna(0) + train_reduced['Pool QC_Gd'].fillna(0) + train_reduced['Pool QC_TA'].fillna(0) + train_reduced['Bsmt Cond_Ex'].fillna(0) + train_reduced['Bsmt Cond_Fa'].fillna(0) + train_reduced['Bsmt Cond_Gd'].fillna(0) + train_reduced['Bsmt Cond_Po'].fillna(0) + train_reduced['Bsmt Cond_TA'].fillna(0)


# Rename the Series to 'LuxRating'
LuxRating = LuxRating.rename('LuxRating')


# Check to see if feature eng is working
print(LuxRating)


# Concatenate the new LuxRating column with the original DataFrame
train_reduced = pd.concat([train_reduced,LuxRating], axis=1)

# Check to see if train now has the new LuxRating column
print(train_reduced)
```

Finally we made this 'Age at Sale' variable, our team hypothesizes that understanding the actual age of the building - based on when it was built and the year it was sold, could be a part of influencing the sales price. Note: Despite that Yr Sold was considered an outlier we thought the variable could still work within the context of Year Built.

# **SIXTH STEP: DECISION TREES**

For our decision trees we chose only to use our OHC and Feature Eng variables that we built having in mind high influence on predicting sales price. We chose to do this for resource efficiency, reduction in standard error, and model simplification.

```python
#We are now choosing to only use our feature eng columns for our decision tree. We chose to do this because several of our variables are combining multiple variables and we have a strong hypothesis that soley these variables can give a well-rounded indication for the predicted sales price.
train_reduced = train_reduced[['TotalSqFt', 'LuxRating','AgeatSale','TotalPorchSF','SalePrice']]

#Check to see if added correctly
train_reduced.head()
```

```python
# We will use your required naming of:
# - X for the data including all our input variables (everything minus SalesPrice)
# - y for the target (SalesPrice)
X = train_reduced.drop(columns=['SalePrice']).fillna(0)
y = train_reduced[['SalePrice']]

#We print the shape to ensure that we minimized the variables correctly.
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

#Before starting our decision tree, we need to build our validation set, so we create a random split for validation
#We use a 30% split as recommended in class

X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.3,random_state=2023)

#We check the shape of the data
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
#import the libraries needed for decision trees
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

result_train_dict = {} # We'll use this to keep track of our training set results
result_val_dict = {} # This one is for our validation set results

# Setting initial maximum values for hyperparameters
max_min_samples_leaf = 20 # Let's limit min_samples_leaf to a maximum of 20
max_max_depth = 15  # Similarly, we'll cap max_depth at 15

# Running through different values of max_depth and min_samples_leaf to start to train our model
for m in range(3, max_max_depth+1, 2): # We vary max_depth from 3 to 15, stepping by 2 each time
    for n in range(1, max_min_samples_leaf+1, 3): # For min_samples_leaf, we'll go from 1 to 20, jumping by 3 each time

# Initializing our DecisionTreeRegressor with our chosen hyperparameters
        tree_reg = DecisionTreeRegressor(random_state=42,
                                         min_samples_leaf=n,
                                         max_depth=m)
 # Fitting the DecisionTreeRegressor model on the training data
        tree_reg.fit(X_train, y_train)

# Making predictions on both the training set and validation set
        train_predicted = tree_reg.predict(X_train)
        val_predicted = tree_reg.predict(X_val)

# Finally we calculate and storing Mean Squared Error (MSE) for training set and validation set
        result_train_dict[m, n] = mean_squared_error(np.log(y_train),
                                                     np.log(train_predicted),
                                                     squared = False)
        result_val_dict[m, n] = mean_squared_error(np.log(y_val),
                                                   np.log(val_predicted),
                                                   squared = False)
```

```python
# Next we plot the different train and validation errors,
# so we can then choose the best value for both max_depth and min_samples_leaf

#importing our needed libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Creating a figure with two subplots (ax1 and ax2) for plotting
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 7))
colors= ['black', 'red', 'blue', 'green', 'orange', 'yellow', 'gray'] # Defining our colors for the plot

# Creating variables for indexing colors and storing custom legend items
color_index = 0
custom_lines = np.array([])
custom_names = np.array([])

# Generating x-axis values for min_samples_leaf
x_axis = np.arange(start=1, stop=max_min_samples_leaf+1, step=3)

# Going through different values of max_depth
for m in range(3, max_max_depth+1, 2):

# Lists to store the y-values (errors) for each value of min_samples_leaf
    line_plot_train = []
    line_plot_val = []

# Going through different values of min_samples_leaf
    for n in x_axis:
        line_plot_train.append(result_train_dict.get((m,n))) # Adding train error for each min_samples_leaf
        line_plot_val.append(result_val_dict.get((m,n))) # Adding the validation error for each min_samples_leaf

# Plotting the train and validation errors for each max_depth on separate subplots
    ax1.plot(x_axis, line_plot_train, alpha=0.5, c=colors[color_index], linestyle='--')
    ax2.plot(x_axis, line_plot_val, alpha=0.5, c=colors[color_index])

 # Creating a custom legend item for each max_depth
    color_line = Line2D([0], [0], color=colors[color_index], lw=4)
    custom_lines = np.append(custom_lines,color_line)
    color_name = 'max_depth =' + str(m)
    custom_names = np.append(custom_names,color_name)
    color_index+=1 # Incrementing our color index for the next max_depth

# Setting x-axis limits and tick marks for both subplots to create efficiency
ax1.set_xlim(max_min_samples_leaf, 0)
ax2.set_xlim(max_min_samples_leaf, 0)
plt.xticks(np.arange(start=1, stop=max_min_samples_leaf, step=2), rotation=90, size=10)

# Adding gridlines to both subplots and labeling axes
ax1.grid(True)
ax2.grid(True)
plt.xlabel('# min_samples_leaf')
plt.ylabel('RMSE')

# Adding a legend with custom_lines and custom_names to the first subplot so we can interpret the graph
ax1.legend(custom_lines, custom_names)

#Displaying our plot
plt.show()
```

```python
# Now we choose different parameters based on the above plots that give more optimized values
# we can fit and predict both train and validation
# and double check that the error values we are getting are similar to the ones in the graph

tree_reg = DecisionTreeRegressor(random_state=42, min_samples_leaf=4, max_depth=15)
tree_reg.fit(X_train, y_train)

# Making predictions on both the training and validation sets
train_predicted = tree_reg.predict(X_train)
val_predicted = tree_reg.predict(X_val)

# Calculating and comparing the Root Mean Squared Error (RMSE) for training and validation sets
(mean_squared_error(np.log(y_train), np.log(train_predicted), squared = False),
 mean_squared_error(np.log(y_val), np.log(val_predicted), squared = False))
```

**Analyzing our validation and train errors:** We note here that our Test standard error is .11 and our validation standard error is .17. We conclude then that we chose good hyperparameters because the standard error within validation is not too large and there is not a large difference between the test and validation errors. Even though the train has a smaller error than validation, it is a marginal difference we conclude it doesn't indicate overfit.

# **SEVENTH STEP: RANDOM FOREST**
*We chose Random Forest due to it's high accuracy, lower overfit risk, and it's ability to in run parallel. We did not chose Gradient Boosting because it cannot run parallel, has even more loss in explanability, and risk of overfit. We outweighed the cons of Gradient Boosting with it's benefit of having some higher accuracy.

```python
# We'll iterate through different number of estimators,
# features, and max_depth.
# train the model and test it using RMSE
# in both train and validations sets:

#importing our needed libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

result_train_dict = {} # Dictionary to store training set results
result_val_dict = {}  # Dictionary to store validation set results

max_max_depth =  14 # Maximum value for max_depth hyperparameter
max_n_estimators = 71 # Maximum value for n_estimators hyperparameter

for m in range(1, max_max_depth, 2): # max depth of trees

    print(m) # Just to know how the training is progressing

    for n in range(1, max_n_estimators, 2): # number of trees
        for s in [0.6, 0.7, 0.8, 0.9]:      # % of samples
            for k in [0.6, 0.7, 0.8, 0.9]:  # % of features

# Initializing the Random Forest Regressor with specified hyperparameters
# We set the random_state to ensure the results can be reproduced
# n_estimators is determining the number of trees in the forest
# max_depth is controlling the maximum depth of each tree in the forest
# max_features is determining the maximum number of features
# max_samples is setting the maximum number of samples used to train each tree
                tree_reg = RandomForestRegressor(random_state= 42,
                                                 n_estimators= n,
                                                 max_depth= m,
                                                 max_features= k,
                                                 bootstrap= True,
                                                 max_samples= s)
# Now we fit the Random Forest Regressor model on the training data
                tree_reg.fit(X_train, np.ravel(y_train))

# Next we make predictions on both the training set and validation set
                train_predicted = tree_reg.predict(X_train)
                val_predicted = tree_reg.predict(X_val)

# And finally we calculate and store our mean sq. error for training set and validation set
# We're taking the set from the target variables and then calculating mean sq. error to get a better sense of error
# We store the results for future reference, along with the optimized given hyperparameters
                result_train_dict[m, n, s, k] = mean_squared_error(np.log(y_train),
                                                                   np.log(train_predicted),
                                                                   squared = False)
                result_val_dict[m, n, s, k] = mean_squared_error(np.log(y_val),
                                                                 np.log(val_predicted), squared = False)
```

```python
# Next we create a DataFrame from our validation set RMSE values
# We label the columns as 'val_rmse'
# And sorting the DataFrame by 'val_rmse' in ascending order so it is easy to see the best hyperparameters for validation

pd.DataFrame.from_dict(result_val_dict, orient='index', columns=['val_rmse']
).sort_values(by='val_rmse',ascending=True)
```

Based off of the above output, we can conclude that the optimized hyperparameters (based on the lowest RMSE) for validation are;

number_estimators = 9
max_depth = 13
max_features = .9
max_samples = .6

```python
#We then use our optimized val_rmse and it's corresponding m,n,k, and s values to see what output we get back in train.
pd.DataFrame.from_dict(result_train_dict, orient='index',
                       columns=['train_rmse']).loc[[(9, 13, 0.9, 0.6)]]
```

*Note: we are pleased to see an even lower Train RMSE but again not marginally lower than the Validation RMSE.*

```python
#Next we plot our graph to see the optimized number of estimators
#importing our needed libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Creating a new plot with specific dimensions for better visualization
fig, ax = plt.subplots(figsize=(15, 7))
# Defining colors for different lines on the plot
colors= ['black', 'red', 'blue', 'green', 'orange', 'gray','navy']


# Initializing variables for indexing colors and storing custom legend items
color_index = 0
custom_lines = np.array([])
custom_names = np.array([])

# Setting values for hyperparameters s and k (we  chose .7 for max_features and .6 for max_samples
# based off of optimizing for the smallest RMSE from our above Random Forest output
s=0.7
k=0.6

# Looping over different values to find our max_depth
for m in range(1, max_max_depth, 2):

# Lists to store the y-values for each value of n_estimators
    line_plot_train = []
    line_plot_val = []

 # Looping over different values of n_estimators
    for n in range(1, max_n_estimators, 2):
        line_plot_train.append(result_train_dict.get((m,n,s,k)))
        line_plot_val.append(result_val_dict.get((m,n,s,k)))


# Plotting the validation error evolution for each max_depth
plt.plot(np.arange(start=1, stop=max_n_estimators, step=2),
         line_plot_val, alpha=0.5, c=colors[color_index])

# Creating a custom legend item for each max_depth
color_line = Line2D([0], [0], color=colors[color_index], lw=4)
custom_lines = np.append(custom_lines, color_line)
color_name = 'max_depth =' + str(m)
custom_names = np.append(custom_names, color_name)
color_index += 1  # Incrementing color index for the next max_depth

# Adding gridlines to the plot and labeling axes
ax.grid(True)
plt.xlabel('# n_estimators')
plt.ylabel('MSE-Log')

# Adding a legend with custom_lines and custom_names
ax.legend(custom_lines, custom_names)

# Displaying the plot
plt.show()
```

**Analyzing Number of Est Graph:** Based on the above plot and finding the 'elbow' of where there is the smallest MSE and then where it levels off, we can chose n_estimators of 10.
